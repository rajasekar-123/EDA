```python

def missing_summary(titanic):
    missing = titanic.isnull().sum()
    percent = (missing / len(titanic)) * 100

    result = pd.DataFrame({
        'MissingCount': missing,
        'Percentage': percent.round(2)
    })

    return result[result['MissingCount'] > 0].sort_values(
        by='Percentage',
        ascending=False
    )

missing_summary(titanic)
```
